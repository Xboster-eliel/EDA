EXTRAYENDO LOS NOMBRES DE LA CARPETA

```python
import os

def leer_nombres_archivos_svg(ruta_carpeta):
    """
    Lee los nombres de los archivos .svg en una carpeta y los devuelve como una lista.

    Args:
        ruta_carpeta: La ruta de la carpeta a escanear.

    Returns:
        Una lista de nombres de archivos .svg en la carpeta especificada.
    """
    nombres_archivos_svg = []
    for nombre_archivo in os.listdir(ruta_carpeta):
        if nombre_archivo.endswith('.svg'):
            nombres_archivos_svg.append(nombre_archivo)
    return nombres_archivos_svg

# Rutas de las carpetas
ruta_ideam = '/content/drive/MyDrive/UNAL/PROYECTO DAGRAM/graficos/BARRAS_Mensuales ideam'
ruta_epm = '/content/drive/MyDrive/UNAL/PROYECTO DAGRAM/graficos/BARRAS_Mensuales EPM'

# Obtener listas desde ambas carpetas
lista_ideam = leer_nombres_archivos_svg(ruta_ideam)
lista_epm = leer_nombres_archivos_svg(ruta_epm)

# Unir listas y eliminar duplicados usando set
lista_archivos_svg = list(set(lista_ideam + lista_epm))

# Imprimir la lista de nombres de archivos .svg
print(lista_archivos_svg)
print('***'*20)

import os

def limpiar_nombre_estacion(nombre_sin_ext):
    """
    Recibe el nombre de archivo (sin extensión) y devuelve
    solo la parte que está después del último guión bajo.
    Ejemplo:
      "Ciclo_Anual_Abejorral (2)"  →  "Abejorral (2)"
    """
    # Buscar la posición del último guión bajo
    idx = nombre_sin_ext.rfind("_")
    if idx != -1:
        return nombre_sin_ext[idx+1:].strip()
    else:
        # Si no hay guión bajo, devuelve el nombre tal cual
        return nombre_sin_ext.strip()

# Ahora, para generar la lista de nombres limpios:
nombres_svg_limpios = [
    limpiar_nombre_estacion(os.path.splitext(nombre)[0])
    for nombre in lista_archivos_svg
]

# Convertir a minúsculas para comparación unificada
nombres_svg = set([n.lower() for n in nombres_svg_limpios])
nombres_svg
```

EXTRAYENDO LOS NOMBRES DEL SHAPEFILE (REGION MAGDALENA)

```python
import geopandas as gpd

# 1. Carga tu shapefile
gdf = gpd.read_file('/content/drive/MyDrive/UNAL/PROYECTO DAGRAM/graficos/eliel prueba/DIVISION DE SHAPE INDIVIDUALES DE (REGIONES Y ESTACIONES IDEAM + EPM)/ideam_region_Región Magdalena.shp')

# 2. Definir la función que limpia cada nombre
def limpiar_nombre_gdf(nombre_con_codigo):
    """
    Recibe un string como "SAN MIGUEL [23050100]" y devuelve "SAN MIGUEL".
    Es decir, elimina el espacio + corchetes con el código.
    """
    # Si no hay corchete, devolvemos tal cual
    if '[' not in nombre_con_codigo:
        return nombre_con_codigo.strip()
    # Separar en la parte antes del corchete
    parte_sin_codigo = nombre_con_codigo.split(' [')[0]
    return parte_sin_codigo.strip()

# 3. Aplicar sobre la columna "NOMBRE" y crear una lista con los nombres limpios
nombres_gdf_limpios = [
    limpiar_nombre_gdf(valor)
    for valor in gdf['NOMBRE']
]

# Convertir a minúsculas para comparación unificada
nombres_gdf = set([n.lower() for n in nombres_gdf_limpios])
nombres_gdf
```

COMPARANDO LOS NOMBRES DE LA CARPETA VS LOS SHAPEFILES

```python
import difflib

# 1. Definimos una función que calcula el “ratio” de similitud entre dos cadenas:
def similitud(a: str, b: str) -> float:
    return difflib.SequenceMatcher(None, a, b).ratio()


# 2. Creamos dos estructuras vacías donde iremos guardando las coincidencias:
high_coincidence = []  # para ratio >= 0.85
some_coincidence = []  # para 0.60 <= ratio < 0.85


# 3. Recorremos cada nombre de nombres_svg y lo comparamos con todos los de nombres_gdf:
for nombre_svg in nombres_svg:
    for nombre_gdf in nombres_gdf:
        ratio = similitud(nombre_svg, nombre_gdf)

        if ratio >= 0.85:
            high_coincidence.append((nombre_svg, nombre_gdf, round(ratio, 3)))
        elif 0.60 <= ratio < 0.85:
            some_coincidence.append((nombre_svg, nombre_gdf, round(ratio, 3)))


# 4. Si solo te interesa una lista única de “nombres que tienen high coincidence”,
#    puedes extraer las tuplas o convertirlas en un set. Por ejemplo:
high_coincidence_set = set((a for a, b, _ in high_coincidence))
some_coincidence_set = set((a for a, b, _ in some_coincidence))


# 5. Veamos los resultados:
print("=== HIGH COINCIDENCE (ratio ≥ 0.85) ===")
for a, b, r in high_coincidence:
    print(f"  • '{a}'  ⟷  '{b}'  =>  {r}")

print("\n=== SOME COINCIDENCE (0.60 ≤ ratio < 0.85) ===")
for a, b, r in some_coincidence:
    print(f"  • '{a}'  ⟷  '{b}'  =>  {r}")

# 6. Si solo quieres las listas finales de nombres sin repetición:
print('***NOMBRES EN LA CARPETA ||||||||||||| NOMBRES DEL SHAPE CARGADO***')
print("\nNombres en high_coincidence_set:", high_coincidence_set)
print("Nombres en some_coincidence_set:", some_coincidence_set)
```
